Bài 1. Xử lý data  (Used Car price)

```python
dataset = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',names=['symboling', 'normalized-losses','make','fuel-type','aspiration','num-of-doors'
                                             ,'body-style','drive-wheels','engine-location','wheel-base','length',
                                             'width','height','curb-weight','engine-type','num-of-cylinders','engine-size',
                                             'fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm',
                                             'city-mpg','highway-mpg','price'],header=None)
```

```python
#dataset.info() #Xem thông tin tổng quan của bảng data
#dataset.head()
#dataset.tail()
dataset.sample(10)
```

```python
dataset # Data xuất hiện những giá trị bị thiếu/sai "?"
```

```python
#Thay giá trị "?" thành NaN 
dataset.replace("?", np.nan, inplace = True)
dataset
```

```python
#In ra giá trị là NaN values 
dataset[dataset['normalized-losses'].isnull()]
```

```python
#check NaN data nếu là NaN trả về True và ngược lại
null_data=dataset.isnull()
#null_data.head()
null_data
```

```python
dataset.isna().sum() #Đếm giá trị NaN ở các cột
```

```python
#In ra mô tả chi tiết cho các cột
for i in null_data.columns.values.tolist():
    print(i)
    print(null_data[i].value_counts())
    print("----------------------------------")
```

```python
#Tính trung bình của các chuỗi (nếu có NaN thì không tính)
print("Average of normalized-losses:","%.2f" % dataset["normalized-losses"].astype("float").mean(axis=0))
print("Average of bore:","%.3f" % dataset['bore'].astype('float').mean(axis=0))
print("Average of horsepower:", "%.3f" % dataset['horsepower'].astype('float').mean(axis=0))
print("Average of peak rpm:", "%.3f" % dataset['peak-rpm'].astype('float').mean(axis=0))
print("Average of stroke:", "%.3f" % dataset['stroke'].astype('float').mean(axis=0))

```

```python
#Thay NaN bằng giá trị trung bình 
dataset["normalized-losses"].replace(np.nan, "%.3f"% 122, inplace=True)
dataset["bore"].replace(np.nan,"%.3f" % dataset['bore'].astype('float').mean(axis=0), inplace=True)
dataset['horsepower'].replace(np.nan,"%.3f" % dataset['horsepower'].astype('float').mean(axis=0), inplace=True)
dataset['peak-rpm'].replace(np.nan, "%.3f" % dataset['peak-rpm'].astype('float').mean(axis=0), inplace=True)
dataset['stroke'].replace(np.nan, "%.3f" % dataset['stroke'].astype('float').mean(axis=0), inplace=True)
```

```python
dataset['num-of-doors'].value_counts() #Đếm các giá trị có trong cột (thường sử dụng cho biến rời rạc)
print("Giá trị phổ biến nhất của biến Num-Of-Doors là :",dataset['num-of-doors'].value_counts().idxmax()) #In ra giá trị có số lượng nhiều nhất (Mode của cột)
```

```python
#Thay giá trị NaN bằng giá trị phổ biến nhất của cột
dataset["num-of-doors"].replace(np.nan, dataset['num-of-doors'].value_counts().idxmax(), inplace=True)
```

```python
#Xóa giá trị Nan trong biến "Price"
dataset.dropna(subset=["price"], axis=0, inplace=True)
```

```python
dataset.reset_index(drop=True,inplace=True)
dataset
```

Bài 2. Xử lý data  (Loan Pay Off)

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
%matplotlib inline
```

```python
#Load data từ link github
df2 = pd.read_csv('https://raw.githubusercontent.com/maxtran1997gmail/Files/main/loan_train.csv')
```

```python
df2.sample(10)
```

```python
df2['effective_date'] = pd.to_datetime(df2['effective_date'])
df2['due_date'] = pd.to_datetime(df2['due_date'])
df2['Realdate'] = df2['due_date'] - df2['effective_date']
df2['Realdate'] = df2['Realdate'].dt.days.astype(str)
df2['Realdate'] = df2['Realdate'].str.replace('days', '')
df2[['Realdate']] = df2[['Realdate']].astype(int)
df2
```

```python
df2.info()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
df2_label = df2[['loan_status','education','Gender']].apply(LabelEncoder().fit_transform) 
df2 = df2.drop(['loan_status','education','Gender'], axis=1)
df2_final= pd.concat([df2, df2_label], axis=1)
df2_final
```

```python
df2_final['loan_status'].value_counts()
```

```python
df2_final.info()
```

```python
!pip install imbalanced-learn
```

```python
y = df2_final[['loan_status']].values
X = df2_final[['Principal','terms', 'education', 'Gender','Realdate']].values
```

```python
from imblearn.under_sampling import NearMiss
NearMiss_Model=NearMiss(version=2)
X_under, y_under=NearMiss_Model.fit_resample(X,y)
X_under.shape, y_under.shape 
```

```python
np.count_nonzero(y_under == 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, test_size=0.3, random_state=10)
from sklearn import tree 
model_tree = tree.DecisionTreeClassifier() 
model_tree.fit(X_train, y_train) 
```

```python
model_tree.score(X_test,y_test)
```

```python
y_pred=model_tree.predict(X_test)
```

```python
import seaborn as sns
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
class_names=['COLLECTION','PAIDOFF']
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Thực tế')
plt.xlabel('Dự báo')
```

```python
import time

start_time = time.time()

from sklearn import tree 
model_tree = tree.DecisionTreeClassifier() 
model_tree.fit(X_train, y_train)

end_time = time.time()

running_time = end_time - start_time

print("Thời gian chạy mô hình: ", running_time, "giây")
```

```python
pip install imbalanced-learn
```

```python
from imblearn.combine import SMOTEENN
X_under, y_under = SMOTEENN().fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, test_size=0.3,random_state=11)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=2)
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
knn_model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
class_names=['1','2','3','4']
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Thực tế')
plt.xlabel('Dự báo')
```
